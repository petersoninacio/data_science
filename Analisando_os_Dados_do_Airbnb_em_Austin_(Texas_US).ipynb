O serviço consiste em conectar a pessoa que oferece a acomodação e a pessoa que quer viajar, facilitando a apresentação e comunicação entre as partes interessadas na locação.

No final de 2018, a Startup fundada 10 anos atrás, já havia **hospedado mais de 300 milhões** de pessoas ao redor de todo o mundo, desafiando as redes hoteleiras tradicionais.

Uma das iniciativas do Airbnb é disponibilizar dados do site sobre algumas das principais cidades do mundo. Por meio do portal [Inside Airbnb](http://insideairbnb.com/get-the-data.html), é possível baixar uma grande quantidade de dados para desenvolver projetos e soluções de *Data Science*.

#### Histogramas sem outliers
 
 Os histogramas abaixo foram tratados e podemos observar uma melhor distribuição dos dados, sem a interferência de pontos fora da curva de análise, que poderiam interferir em nossa analise.

###### 2015

```
# remover os outliers em um novo DataFrame
df_clean_2015 = df_2015.copy()
df_clean_2015.drop(df_clean_2015[df_clean_2015.minimum_nights > 60].index, axis=0, inplace=True)
df_clean_2015.drop(df_clean_2015[df_clean_2015.price > 1600].index, axis=0, inplace=True)

# remover neighbourhood_group pois esta vazio
df_clean_2015.drop('neighbourhood_group', axis=1, inplace=True)

# plotar o histograma para as variáveis numéricas
df_clean_2015.hist(bins=50, figsize=(15, 10));
```

###### 2020

```
# remover os outliers em um novo DataFrame
df_clean_2020 = df_2020.copy()
df_clean_2020.drop(df_clean_2020[df_clean_2020.minimum_nights > 60].index, axis=0, inplace=True)
df_clean_2020.drop(df_clean_2020[df_clean_2020.price > 1600].index, axis=0, inplace=True)

# remover neighbourhood_group pois esta vazio
df_clean_2020.drop('neighbourhood_group', axis=1, inplace=True)

# plotar o histograma para as variáveis numéricas
df_clean_2020.hist(bins=50, figsize=(15, 10));
```

## Conclusões


A cidade de Austin(TX), apresentou um crescimento intenso nos últimos anos, graças ao forte apelo de ser uma cidade propícia ao empreendedorismo e um centro reconhecido para startups.

Um dos vértices que podem ser analisados quando um local encontra-se em franco crescimento é sua capacidade de receber um número crescente de pessoas. Como esta é uma análise inicial, vamos considerar somente 2 pontos que podem comprovar esse crescimento. Um deles é a quantidade de acomodações e outro é a disponibilidade dos imóveis.

Através de dados fornecidos pelo Airbnb, e uma diferença de 5 anos no espaço amostral(2015 e 2020) podemos verificar que a quantidade de acomodações aumentou em aproximadamente 117%, saindo de 5192 para 11282 acomodações. Isso não garante que tivemos um crescimento populacional, mas é um indício que um aumento demográfico aconteceu ou que houve uma maior credibilidade ao Airbnb.

Se somarmos a crecente oferta de acomodações e a diminuição na disponibilidade destas, podemos concluir que um número maior de pessoas procuram se hospedar na cidade.Pode ser em busca de turismo ou de negócios, ou ainda para apreciar as qualidades da cidades como futuro destino habitacional. Portanto, pode-se verificar que a disponibilidade das acomodações caiu de 72% em 2015 para 32% em 2020.

Com um aumento na quantidade de acomodações e diminuição na disponibilidade, verificamos que um número maior de pessoas se hospedam na cidade. O número mínimo de permanência também aumentou, de 1,8 para 4 dias.

Não se pode afirmar com toda certeza que somente com uma fonte de informação é suficiente para concluírmos tal crescimento. Seria interessante analisar outras fontes, como a quantidade de acomodações e disponibilidade na rede hoteleira da cidade.
